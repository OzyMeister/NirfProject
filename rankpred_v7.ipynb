```python
df = data.fillna(0)
df
```

```python
data
```

```python
for i in range(183): 
  cnt=0
  if(df._get_value(i,'a')!=0):
    cnt+=1
  if(df._get_value(i,'b')!=0):
    cnt+=1
  if(df._get_value(i,'c')!=0):
    cnt+=1
  if(df._get_value(i,'d')!=0):
    cnt+=1
  if(df._get_value(i,'e')!=0):
    cnt+=1
  
  

  new=(df._get_value(i,'a')+df._get_value(i,'b')+df._get_value(i,'c')+df._get_value(i,'d')+df._get_value(i,'e'))/cnt
  data.loc[i] = data.loc[i].fillna(new)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(df, test_size=0.3, random_state=1, shuffle=True)
```

```python
isinstance(train, pd.DataFrame)
```

```python
test
```

```python
train
```

```python
frames = [train,train,train,train,train,train,train] #8x oversampling for knn
train = pd.concat(frames)
```

```python
train
```

```python
x=train.drop(['Y'],axis=1).values
y=train['Y'].values
```

```python
x_test=test.drop(['Y'],axis=1).values
y_test=test['Y'].values
```

```python
#from sklearn.model_selection import train_test_split
#x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3, random_state=1)
```

```python
from sklearn.linear_model import LinearRegression
ml=LinearRegression()
ml.fit(x,y)
ml.score(x_test,y_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
neigh = KNeighborsRegressor(n_neighbors=2)
neigh.fit(x,y)
neigh.score(x_test,y_test)
```

```python
y_pred=neigh.predict(x_test)
print(y_pred)
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(ml,x,y,scoring='r2', cv=5)
scores
```

```python
from sklearn.linear_model import Ridge
rr=Ridge()
rr.fit(x,y)
rr.score(x_test,y_test)
```

```python
from sklearn.neighbors import KNeighborsRegressor
neigh = KNeighborsRegressor(n_neighbors=2)
neigh.fit(x,y)
neigh.score(x_test,y_test)
```

```python
from sklearn import linear_model
clf = linear_model.Lasso(alpha=0.1)
clf.fit(x,y)
clf.score(x_test,y_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
regr = RandomForestRegressor(max_depth=4, random_state=1)
regr.fit(x,y)
regr.score(x_test,y_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor = DecisionTreeRegressor(random_state=0)
regressor.fit(x,y)
regressor.score(x_test,y_test)
```

```python
import pandas as pd
import numpy as np
from keras.models import Sequential
from keras.layers import Dense
```

```python
# create a Sequential model
model = Sequential()

# add input layer
model.add(Dense(32, input_dim=5, activation='relu'))

# add hidden layer
model.add(Dense(16, activation='relu'))

# add output layer
model.add(Dense(1, activation='sigmoid'))

# compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# train the model
model.fit(x, y, epochs=100, batch_size=32, validation_split=0.2)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# create a Sequential model
model = Sequential()

# add input layer
model.add(Dense(units=32, activation='relu', input_dim=5))

# add hidden layers
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=128, activation='relu'))

# add output layer
model.add(Dense(units=1, activation='sigmoid'))

# compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# train the model
model.fit(x, y, epochs=100, batch_size=32, validation_data=(x_test, y_test))
```
