```python
import tabula
import pandas as pd
import numpy as np

# Set the path to your PDF file
pdf_path = r'C:\Users\Astik\Downloads\IR-E-U-0573.pdf'

# Use tabula to extract tables from the PDF file
tables = tabula.read_pdf(pdf_path, pages="all")

# Convert each table into a Pandas DataFrame
dfs = []
for i, table in enumerate(tables, start=1):
    df = pd.DataFrame(table)
    df.columns = [f"{col} T{i}_C{j+1}" for j, col in enumerate(df.columns)]
    dfs.append(df)

# Display the resulting DataFrames
for i, df in enumerate(dfs, start=1):
    print(f"Table {i}")
    display(df)
    print()
```

```python
print(f"Number of DataFrames: {len(dfs)}")
```

```python
dfs[0] = dfs[0].drop(dfs[0].columns[0], axis=1)


# assume dfs is a list of dataframes
for i in range(len(dfs)):
    dfs[i] = dfs[i].replace('-', np.nan)
    dfs[i] = dfs[i].fillna('0')
```
